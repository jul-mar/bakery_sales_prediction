```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../0_DataPreparation/00_data/data_long.csv")
df
```

```python
df_subset = df.loc[1819:(1818*2), ['Datum', 'Umsatz', 'VPI']]
df_subset
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_subset['VPI'], df_subset['Umsatz'], alpha=0.7, color='blue')
plt.title('Zusammenhang zwischen VPI und Umsatz')
plt.xlabel('VPI (normalisiert)')
plt.ylabel('Umsatz')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_subset['Datum'], df_subset['Umsatz'], label='Umsatz', color='blue')
plt.plot(df_subset['Datum'], df_subset['VPI'] * df_subset['Umsatz'].max(), label='VPI (skaliert)', color='orange')  # VPI skaliert
plt.title('Umsatz und VPI über die Zeit')
plt.xlabel('Datum')
plt.ylabel('Wert')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
