```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np

data = pd.read_csv("/workspaces/bakery_sales_prediction/0_DataPreparation/Trainingsdaten.csv")

#erste drei Testmodelle
model_1 = smf.ols('Brot ~ Umsatz_total', data=data).fit()
model_2 = smf.ols('Broetchen ~ Umsatz_total', data=data).fit()
model_3 = smf.ols('Umsatz_total ~ Broetchen', data=data).fit()
model_4 = smf.ols('Brot ~ Broetchen', data=data).fit()

# Display the summary of the initial model
#model_1.summary()

# R^2 und p Werte einzeln ausgeben, damit man nicht immer direkt diesen riesigen summary Block als Ausgabe bekommt
print("Brot ~ Gesamtumsatz:")
print(f"adj. R^2: {model_1.rsquared_adj:.4f}")
print("P-Werte:")
p_values = model_1.pvalues
for var, p_val in p_values.items():
            print(f"{var}: {p_val:.4f}")
            
print("\nBrötchen ~ Gesamtumsatz:")
print(f"adj. R^2: {model_2.rsquared_adj:.4f}")
print("P-Werte:")
p_values = model_2.pvalues
for var, p_val in p_values.items():
            print(f"{var}: {p_val:.4f}")
            
print("\nGesamtumsatz ~ Brötchen:")
print(f"adj. R^2: {model_3.rsquared_adj:.4f}")
print("P-Werte:")
p_values = model_3.pvalues
for var, p_val in p_values.items():
            print(f"{var}: {p_val:.4f}")
            
print("\nBrot ~ Brötchen:")
print(f"adj. R^2: {model_4.rsquared_adj:.4f}")
print("P-Werte:")
p_values = model_4.pvalues
for var, p_val in p_values.items():
            print(f"{var}: {p_val:.4f}")
```

```python
# brute force "alles" durchtesten
# Funktion zur Durchführung und Vergleich von Regressionsmodellen
def compare_regression_models(data):
    # Liste der Modelle, die wir testen werden
    models = {
        # Modelle mit einer Variablen
        'Brot ~ Umsatz_total': smf.ols('Brot ~ Umsatz_total', data=data).fit(),
        'Brötchen ~ Umsatz_total': smf.ols('Broetchen ~ Umsatz_total', data=data).fit(),
        'Brot ~ Brötchen': smf.ols('Brot ~ Broetchen', data=data).fit(),
        'Umsatz_total ~ Heimspiel':  smf.ols('Umsatz_total ~ Heimspiel', data=data).fit(),
        'Umsatz_total ~ Weihnachtsmarkt':  smf.ols('Umsatz_total ~ Weihnachtsmarkt', data=data).fit(),
        'Umsatz_total ~ Markt':  smf.ols('Umsatz_total ~ Markt', data=data).fit(),
        'Brot ~ Heimspiel':  smf.ols('Brot ~ Heimspiel', data=data).fit(),
        'Brot ~ Weihnachtsmarkt':  smf.ols('Brot ~ Weihnachtsmarkt', data=data).fit(),
        'Brot ~ Markt':  smf.ols('Brot ~ Markt', data=data).fit(),
        'Brot ~ Weather_Good':  smf.ols('Brot ~ Weather_Good', data=data).fit(),
        'Brot ~ Feiertag':  smf.ols('Brot ~ Feiertag', data=data).fit(),
        'Brot ~ Montag': smf.ols('Brot ~ Montag', data=data).fit(),
        'Brot ~ Dienstag': smf.ols('Brot ~ Dienstag', data=data).fit(),
        'Brot ~ Mittwoch': smf.ols('Brot ~ Mittwoch', data=data).fit(),
        'Brot ~ Donnerstag': smf.ols('Brot ~ Donnerstag', data=data).fit(),
        'Brot ~ Freitag': smf.ols('Brot ~ Freitag', data=data).fit(),
        'Brot ~ Samstag': smf.ols('Brot ~ Samstag', data=data).fit(),
        'Brot ~ Sonntag': smf.ols('Brot ~ Sonntag', data=data).fit(),
        'Brötchen ~ Heimspiel':  smf.ols('Broetchen ~ Heimspiel', data=data).fit(),
        'Brötchen ~ Weihnachtsmarkt':  smf.ols('Broetchen ~ Weihnachtsmarkt', data=data).fit(),
        'Brötchen ~ Markt':  smf.ols('Broetchen ~ Markt', data=data).fit(),
        
        # Modelle mit 2 Variablen
        'Brot ~ Umsatz_total + Heimspiel': smf.ols('Brot ~ Umsatz_total + Heimspiel', data=data).fit(),
        'Brot ~ Umsatz_total + Weihnachtsmarkt': smf.ols('Brot ~ Umsatz_total + Weihnachtsmarkt', data=data).fit(),
        'Umsatz_total ~ Brot + Weihnachtsmarkt': smf.ols('Umsatz_total ~ Brot + Weihnachtsmarkt', data=data).fit(),
        'Brot ~ Umsatz_total + Markt': smf.ols('Brot ~ Umsatz_total + Markt', data=data).fit(),
        'Brötchen ~ Umsatz_total + Heimspiel': smf.ols('Broetchen ~ Umsatz_total + Heimspiel', data=data).fit(),
        'Brötchen ~ Umsatz_total + Weihnachtsmarkt': smf.ols('Broetchen ~ Umsatz_total + Weihnachtsmarkt', data=data).fit(),
        'Umsatz total ~ Brötchen + Weihnachtsmarkt': smf.ols('Umsatz_total ~ Broetchen + Weihnachtsmarkt', data=data).fit(),
        
        # Komplexeres Modell mit mehreren Variablen
        'Brot ~ Umsatz_total + Heimspiel + Weihnachtsmarkt + Markt': 
            smf.ols('Brot ~ Umsatz_total + Heimspiel + Weihnachtsmarkt + Markt', data=data).fit(),
        'Brot ~ Montag + Dienstag + Mittwoch + Donnerstag + Freitag + Samstag + Sonntag': 
            smf.ols('Brot ~ Montag + Dienstag + Mittwoch + Donnerstag + Freitag + Samstag + Sonntag', data=data).fit()  
    }
    
    # Ausgabe der adjustierten R²-Werte und p-Werte
    print("Modellvergleich:")
    for name, model in models.items():
        print(f"\nModell: {name}")
        print(f"Adjustiertes R-Quadrat: {model.rsquared_adj:.4f}")
        
        # Ausgabe der p-Werte
        print("P-Werte:")
        p_values = model.pvalues
        for var, p_val in p_values.items():
            print(f"{var}: {p_val:.4f}")
    
    # Finden des Modells mit dem höchsten adjustierten R²
    best_model_name = max(models, key=lambda x: models[x].rsquared_adj)
    best_model = models[best_model_name]
    
    print("\n--- Bestes Modell ---")
    print(f"Modellgleichung: {best_model_name}")
    print(f"Adjustiertes R-Quadrat: {best_model.rsquared_adj:.4f}")
    
    # Detaillierte Zusammenfassung des besten Modells
    print("\nDetaillierte Zusammenfassung des besten Modells:")
    print(best_model.summary())
    
    return best_model, best_model_name

# Hauptausführung
if __name__ == "__main__":
    # Modellvergleich durchführen
    best_model, best_model_name = compare_regression_models(data)
```
