### interaktive Scatterplots f√ºr die Umsatzdaten um Trends und besonders umsatzstarke Tage zu finden

```python

import pandas as pd
import plotly.express as px

```

```python
# Datensatz erstellen

df = pd.read_csv("/workspaces/bakery_sales_prediction/0_DataPreparation/umsatzdaten_gekuerzt.csv")
print(df.head())

```

```python
# Sicherstellen, dass die Datenspalte als datetime interpretiert wird
df["Datum"] = pd.to_datetime(df["Datum"])

# Jahr aus dem Datum extrahieren
df["Jahr"] = df["Datum"].dt.year

# Plots direkt im Jupyter Notebook anzeigen
for jahr in df["Jahr"].unique():
    for warengruppe in df["Warengruppe"].unique():
        subset = df[(df["Jahr"] == jahr) & (df["Warengruppe"] == warengruppe)]
        if not subset.empty:
            fig = px.scatter(
                subset,
                x="Datum",
                y="Umsatz",
                title=f"Scatterplot f√ºr Jahr {jahr}, Warengruppe {warengruppe}",
                labels={"Datum": "Datum", "Umsatz": "Umsatz (EUR)"},
                color="Warengruppe",
            )
            display(fig)  # Plot im Notebook anzeigen
```
